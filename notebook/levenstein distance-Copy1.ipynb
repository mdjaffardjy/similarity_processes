In this notebook we'll compute the levenstein distance for all pairs of processes

```python
def levenshtein_snk(snk_rules, resume=0, output_file = "/home/marinedjaffardjy/Documents/Code/Similarite_process/json/levenshtein_tools_snk.json"):
    v1 = snk_rules.copy()[resume:-1]
    v2 = snk_rules.copy()[resume+1:]
    i = 0
    if(resume>0):
        with open(output_file) as f:
            scores = json.load(f)
    else :
        scores = []
    for rule1 in v1:
        print(str(i)+"/"+str(len(v1)))
        i+=1
        for rule2 in v2:
            score = jellyfish.levenshtein_distance(rule1['code'],rule2['code'])
            l = max(len(rule1['code']),len(rule2['code']))
            scores.append({"rule1":rule1["wf_orig"]+"/"+rule1["name"],
                           "rule2":rule2["wf_orig"]+"/"+rule2["name"],
                           "levenshtein":(l-score)/l})
        if(len(v2)>=2):
            v2 = v2[1:]
        with open(output_file,"w") as f:
            json.dump(scores,f)
        output_file
    return scores
```
